Create dataset folder structure

```python
import os

os.makedirs('dataset/train/normal', exist_ok=True)
os.makedirs('dataset/train/dr', exist_ok=True)
os.makedirs('dataset/val/normal', exist_ok=True)
os.makedirs('dataset/val/dr', exist_ok=True)
```

Import necessary libraries

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np
import os
from tensorflow.keras.preprocessing import image
```

```python
Define image size and data generators for loading and augmenting images
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    horizontal_flip=True,
    fill_mode="nearest"
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    'dataset/train',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    'dataset/val',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)
```

Build Convolutional Neural Network (CNN) Model

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    BatchNormalization(),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

Train the model on the training data with validation

```python
EPOCHS = 25

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator
)
```

Evaluate model performance on validation set

```python
val_generator.reset()
pred_probs = model.predict(val_generator)
pred_classes = (pred_probs > 0.5).astype(int).reshape(-1)
true_classes = val_generator.classes

print("Classification Report:")
print(classification_report(true_classes, pred_classes))

print("Confusion Matrix:")
print(confusion_matrix(true_classes, pred_classes))
```

Plot ROC curve to evaluate classification threshold performance

```python
auc_score = roc_auc_score(true_classes, pred_probs)
fpr, tpr, _ = roc_curve(true_classes, pred_probs)

plt.figure()
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

 Load validation images to visualize true vs predicted labels

```python
def load_images_from_folder(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(folder, filename)
            img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
            img_array = image.img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(label)
    return images, labels

normal_images, normal_labels = load_images_from_folder('dataset/val/normal', 0)
dr_images, dr_labels = load_images_from_folder('dataset/val/dr', 1)

val_images = normal_images + dr_images
val_labels = normal_labels + dr_labels

for i in range(len(val_images)):
    img = np.expand_dims(val_images[i], axis=0)
    pred_prob = model.predict(img)[0][0]
    pred_label = 'DR' if pred_prob > 0.5 else 'Normal'
    true_label = 'DR' if val_labels[i] == 1 else 'Normal'

    plt.imshow(val_images[i])
    plt.title(f"True: {true_label} - Pred: {pred_label} ({pred_prob:.2f})")
    plt.axis('off')
    plt.show()
```
